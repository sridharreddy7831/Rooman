# Data Selection & Filtering

```python
a.head()
```

```python
# selecting the particular column and getting the 5 values by using the head()

a['Age'].head()
```

```python
# Selecting the Multiple Columns 

a[['BloodPressure','BMI']]
```

```python
# Selecting the Multiple Columns of first 6 rows 

a[['BloodPressure','BMI']].head(6)
```

```python
# Selecting the Multiple Columns of last 7 rows 

a[['BloodPressure','BMI']].tail(7)
```

```python
a.iloc[2:5,2:7] # it will select the rows from 2 to 5 and columns from 2 to 7
```

```python
a.loc[2:5,['Age','BMI']] # it will select the rows from 2 to 5 and columns from Age and BMI
```

```python
a.loc[2:5] # it will select the rows from 2 to 5 and all the columns
```

```python
a[a['Age'] < 30] # it will select the rows which are lessthan than 30 in the Age column
```

```python
# it will select the rows which are lessthan than 30 in the Age column and displays the Age and BMI columns of first 10 values

a[a['Age'] < 30][['Age','BMI']].head(10) 
```

```python
# it will select the rows which are lessthan than 30 in the Age column

a.query('Age < 30').head(10) # it will select the rows which are lessthan than 30 in the Age column and displays the Age and BMI columns of first 10 values
```

```python
# it will select the value of the particular cell by using the row and column name

a.at[2,'Age'] 
```

```python
a.iat[2,1] # it will select the value of the particular cell by using the row and column index
```
